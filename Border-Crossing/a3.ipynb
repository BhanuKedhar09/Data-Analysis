<h3> Q.2) Largest Pedestrian Crossing</h3>

```python
count = 0
d = dict()
for j in data:
    if j["Measure"] == "Pedestrians":
        if count < j["Value"]:
            count = j["Value"]
            d.update({"Value" : count, "Date" : j["Date"], "Port Name" : j["Port Name"], "State" : j["State"] })
d
```

<h3>Q.3.a) Duplicate Port Name</h3>

```python
d1 = dict()
d2 = list()
lis = list()
for k in data:
    if k["Port Code"] not in d1:
        d1[k["Port Code"]] =  k["Port Name"]
        d2.append({k["Port Name"] : k["State"]})
        lis.append(k["Port Name"])
res = {x for x in lis if lis.count(x) > 1}
result = "".join(list(res))
se =list()
for key, value in d1.items():
    if value == result:
        se.append({"portcode" : key , "portname" : value, "state" : None})
count = 0
for j in range(len(d2)):
    for k,v in d2[j].items():
        if k == result:
            se[count]["state"] = v
            count += 1
se
```

<h3>Q4) Reformat the Data</h3>

```python
temp = dict()
for i in data:
    t = (i["Port Code"], i["Date"])
    if t not in temp.keys():
        k = {'Port Name' : i['Port Name'], 'State' : i['State'], 
              'Port Code' : i['Port Code'], 'Border' : i['Border'], 
              'Date' : i['Date'], i['Measure'] : i['Value']}
        temp[t] = k
    else :
        y = {i['Measure'] : i['Value']}
        temp[t].update(y)
len(temp)
```

```python
new_data = list(temp.values())
new_data
```

<h3>Q5) Compute and Add Total Number of Persons</h3>

```python
for i in new_data:
    total = 0
    if "Bus Passengers" in i.keys():
        total = total + i["Bus Passengers"]
    if 'Pedestrians' in i.keys():
        total = total + i['Pedestrians']
    if 'Personal Vehicle Passengers' in i.keys():
        total = total + i['Personal Vehicle Passengers']
    if 'Train Passengers' in i.keys():
        total = total + i['Train Passengers']
    i.update({'Total People' : total})
new_data
```

<h3>Q6) Filter by Transportation Mode</h3>

```python
def count_by_mode(transport_mode):
    t = set()
    for port in new_data:
        if transport_mode in port.keys() and port[transport_mode] > 0 :
            t.add(port["Port Code"])
    return len(t)
print(count_by_mode('Trains'))
count_by_mode('Personal Vehicles')
```
