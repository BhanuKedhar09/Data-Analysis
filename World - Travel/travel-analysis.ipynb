```python
# approch :
# i noticed the data from world bank and unwto are having similar records
# so i took the common countries from both world bank and unwto
# grouped both wb and unwto by countrycodealpha to seperate the rows based on countries
# now merged same countries, after merging the indicater shows if the record is present in both or right_only or left_only
# only taking records that are left_only and right_only
# and concatinating all of the uncommon records
# that means if the wb has 8 record and unwto has 10 records(for each country) and every wb record or most of the wb records are in unwto
# the uncommon ones are going to be from the unwto
```

```python
wb_countries = set(wb["countryCodeAlpha"].unique())
unwto_countries = set(unwto["countryCodeAlpha"].unique())
common_countries = wb_countries & unwto_countries

# converting to int to avoid problems
wb["receipts"] = wb["receipts"].fillna(0).astype("int")
wb["expenditures"] = wb["expenditures"].fillna(0).astype("int")

unwto["receipts"] = unwto["receipts"].fillna(0).astype("int")
unwto["expenditures"] = unwto["expenditures"].fillna(0).astype("int")

grouped_ec = wb.groupby("countryCodeAlpha")
grouped_ec1 = unwto.groupby("countryCodeAlpha")

con = []

for cc in common_countries:
    # print(~(grouped_ec.get_group(cc).reset_index(drop = True)).isin(grouped_ec1.get_group(cc).reset_index(drop = True)).all(axis=1))
    df1 = grouped_ec.get_group(cc).reset_index(drop = True)
    df2 = grouped_ec1.get_group(cc).reset_index(drop = True)
    merged = pd.merge(df1, df2, how='outer', indicator=True)
    df_diff = merged.loc[merged['_merge'] != 'both']
    df_diff["_merge"] = df_diff["_merge"].str.replace("left_only", "wb")
    df_diff["_merge"] = df_diff["_merge"].str.replace("right_only", "unwto")
    con.append(df_diff)

con = pd.concat(con)
con
```
