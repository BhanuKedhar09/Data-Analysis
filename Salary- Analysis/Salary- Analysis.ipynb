## 1. Reformatting the Salary column

```python
df["Salary"] = (df["Salary"].str.replace(",", "")).astype("float")
```

```python
df.head()
```

## 2. Updating the Range Columns

```python
df["Age"].unique(), df["ExpOverall"].unique(), df["ExpInField"].unique()
```

```python
# df["Age"] = df["Age"].str.replace("under ", "")
df.loc[df["Age"].str.contains("under "), "Age"] = "0-17"
df.loc[df["Age"].str.contains("over"), "Age"] = "65-999"
df.loc[df["ExpOverall"].str.contains("less"), "ExpOverall"] = "0 - 1 years"
df.loc[df["ExpOverall"].str.contains("more"), "ExpOverall"] = "41 - 999 years"
df.loc[df["ExpInField"].str.contains("less"), "ExpInField"] = "0 - 1 years"
df.loc[df["ExpInField"].str.contains("more"), "ExpInField"] = "41 - 999 years"
```

```python
df["Age"].unique(), df["ExpOverall"].unique(), df["ExpInField"].unique()
```

```python
df[["Start_Age", "End_Age"]] = df["Age"].str.split("-", expand=True)
df["Start_Age"] = df["Start_Age"].astype("int32")
df["End_Age"] = df["End_Age"].astype("int32")
df["End_Age"] = df["End_Age"] + 1
```

```python
df["Age"] = pd.IntervalIndex.from_arrays(df['Start_Age'], df['End_Age'], closed='left')
```

```python
df.head()
df[df.Age.array.overlaps(pd.Interval(18,45, closed='left'))]
```

```python
len(df)
```

```python
df["ExpOverall"] = df["ExpOverall"].str.replace("years", "")
df[["Start_expo", "End_expo"]] = df["ExpOverall"].str.split("-", expand=True)
df["Start_expo"] = df["Start_expo"].str.strip()
df["End_expo"] = df["End_expo"].str.strip()
df["Start_expo"] = df["Start_expo"].astype("int32")
df["End_expo"] = df["End_expo"].astype("int32")
df["End_expo"] = df["End_expo"] + 1
```

```python
df["ExpOverall"] = pd.IntervalIndex.from_arrays(df['Start_expo'], df['End_expo'], closed='left')
df.head()
```

```python
df["ExpInField"] = df["ExpInField"].str.replace("years", "")
df[["Start_expf", "End_expf"]] = df["ExpInField"].str.split("-", expand=True)
df["Start_expf"] = df["Start_expf"].str.strip()
df["End_expf"] = df["End_expf"].str.strip()
df["Start_expf"] = df["Start_expf"].astype("int32")
df["End_expf"] = df["End_expf"].astype("int32")
df["End_expf"] = df["End_expf"] + 1
```

```python
df["ExpInField"] = pd.IntervalIndex.from_arrays(df['Start_expf'], df['End_expf'], closed='left')
df.head()
```

## 5. Salary by State

```python
top_five = df['Industry'].value_counts().head(5).index
```

```python
df = df[df['Industry'].isin(top_five)]
```

```python
df = df.assign(State=df['State'].str.split(', ')).explode('State')
```

```python
len(df)
```

```python
df = df.dropna(subset=['Salary'])
```

```python
df_avg = df.groupby(['Industry', 'State'])['Salary'].mean().reset_index()
```

```python
df_avg
```

```python
df_pivot = df_avg.pivot(index='State', columns='Industry', values='Salary')
```
